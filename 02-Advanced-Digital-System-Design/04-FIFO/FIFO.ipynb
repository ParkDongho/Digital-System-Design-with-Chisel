# Skid Buffer

```scala
class SkidBuffer(width: Int) extends FIFO(1, width){
  io.enq.ready := ~ io.deq.valid | io.deq.ready
  io.deq.bits  := RegEnable(io.enq.bits,  io.enq.ready)
  io.deq.valid := RegEnable(io.enq.valid, io.enq.ready)
}

visualize(() => new SkidBuffer(32))
println(getVerilog(new SkidBuffer(32)))
```

# Shift FIFO

```scala
class ShiftFIFO(depth: Int, width: Int) extends FIFO{
  
  //Module Instantiation (Counter)
  val counter = new MyCounter(depth-1) // 0 to depth-1
  //Module Instantiation (ShiftMem)
  io.deq.bits := ShiftMem(io.enq.bits, counter.value.asUInt, io.enq.fire)
  
  //Status(Counter)
  io.enq.ready :=  counter.isMax()
  io.deq.valid := ~counter.isMin()

  //Control
  when(io.enq.fire ^ io.deq.fire){
    counter.en := true.B  
    when(io.enq.fire){
      counter.inc()
    }
    when(io.deq.fire){
      counter.dec()
    }
  } .otherwise{
    counter.en := false.B
  }
}

object ShiftFIFO{
  def apply(){
    
  }
}

visualize(() => new ShiftFIFO(32))
println(getVerilog(new ShiftFIFO(32)))
```
